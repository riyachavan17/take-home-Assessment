```python
import os
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

def preprocess_data(ghi_root, pr_root, output_file="full_merged_data.csv"):
    all_data = []

    months = os.listdir(ghi_root)
    for month in months:
        ghi_month_path = os.path.join(ghi_root, month)
        pr_month_path = os.path.join(pr_root, month)

        if not os.path.isdir(ghi_month_path) or not os.path.isdir(pr_month_path):
            continue

        daily_files = os.listdir(ghi_month_path)

        for file_name in daily_files:
            ghi_file = os.path.join(ghi_month_path, file_name)
            pr_file = os.path.join(pr_month_path, file_name)

            if not os.path.isfile(ghi_file) or not os.path.isfile(pr_file):
                continue

            try:
                df1 = pd.read_csv(ghi_file)
                df2 = pd.read_csv(pr_file)

                df1["Date"] = pd.to_datetime(df1["Date"], dayfirst=True)
                df2["Date"] = pd.to_datetime(df2["Date"], dayfirst=True)

                df1 = df1[["Date", "GHI"]]
                df2 = df2[["Date", "PR"]]

                merged = pd.merge(df1, df2, on="Date", how="inner")
                all_data.append(merged)

            except Exception as e:
                print(f"❌ Error processing {file_name}: {e}")

    if all_data:
        full_df = pd.concat(all_data, ignore_index=True)
        full_df = full_df.sort_values("Date")
        full_df.to_csv(output_file, index=False)
        print(f"✅ Saved merged data as {output_file} with {len(full_df)} rows.")
        return full_df
    else:
        print("⚠ No data merged.")
        return pd.DataFrame()

# Example usage
ghi_root = r"C:\Users\riyac\Downloads\data\data\GHI"
pr_root = r"C:\Users\riyac\Downloads\data\data\PR"
df = preprocess_data(ghi_root, pr_root)
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from datetime import datetime
import numpy as np

def generate_pr_graph(df, output_image="pr_graph.png"):
    df = df.copy()
    df = df.sort_values("Date")
    df["Date"] = pd.to_datetime(df["Date"])
    df["PR_MA_30"] = df["PR"].rolling(window=30).mean()

    # Budget PR Calculation (starts July 2019)
    start_date = pd.to_datetime("2019-07-01")
    df["Budget_PR"] = 73.9  # Default

    for i in range(len(df)):
        date = df.iloc[i]["Date"]
        year_diff = ((date.year - 2019) * 12 + (date.month - 7)) // 12
        df.at[i, "Budget_PR"] = round(73.9 * (0.992) ** year_diff, 1)

    # Color Mapping for GHI
    def ghi_color(ghi):
        if ghi < 2:
            return "navy"
        elif ghi < 4:
            return "deepskyblue"
        elif ghi < 6:
            return "orange"
        else:
            return "brown"

    df["Color"] = df["GHI"].apply(ghi_color)

    # Plotting
    plt.figure(figsize=(16, 8))
    sns.set(style="whitegrid")

    # Scatter plot
    plt.scatter(df["Date"], df["PR"], c=df["Color"], label="PR (by GHI)", alpha=0.7)

    # Red line – 30-day moving average of PR
    plt.plot(df["Date"], df["PR_MA_30"], color="red", linewidth=2, label="30-day PR Avg")

    # Green line – Budget PR line
    plt.plot(df["Date"], df["Budget_PR"], color="darkgreen", linestyle="--", linewidth=2, label="Budget PR")

    # PR Summary Box
    end_date = df["Date"].max()
    pr_summary = ""
    for days in [7, 30, 60, 90]:
        window = df[df["Date"] >= (end_date - pd.Timedelta(days=days))]
        avg = window["PR"].mean()
        pr_summary += f"Last {days}d PR: {avg:.2f}\n"

    plt.gca().text(1.01, 0.5, pr_summary, transform=plt.gca().transAxes,
                   fontsize=12, verticalalignment='center', bbox=dict(boxstyle="round", facecolor='lightyellow'))

    plt.title("PR Trend with GHI Coloring, 30-day Avg & Budget Line", fontsize=16)
    plt.xlabel("Date")
    plt.ylabel("Performance Ratio (PR)")
    plt.legend()
    plt.tight_layout()
    plt.savefig(output_image)
    plt.show()

    print(f"✅ Graph saved as {output_image}")
```
